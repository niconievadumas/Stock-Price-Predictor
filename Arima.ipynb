```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from datetime import date


```

```python
# Load the stock data into a pandas DataFrame
df = pd.read_csv('NVDA.csv')

# Convert the date column to a datetime object and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df.index = pd.DatetimeIndex(df.index).to_period('D')

df
```

```python
# Fit an ARIMA model to the data
model = sm.tsa.ARIMA(df['Close'], order=(1,0,0))
model_fit = model.fit()

# Make predictions
future_dates = pd.date_range(date.today(), periods=10, freq='D')
# print(future_dates)
# future_prices = model_fit.predict(start=future_dates[0], end=future_dates[-1])
df['Pronostico'] = model_fit.fittedvalues  


# Print the predictions
# print(future_prices)
df
```

```python
modelo = sm.tsa.ARIMA(df['Close'].iloc[1:], order=(2, 1, 0))  
resultados = modelo.fit()  

df['Pronostico'] = resultados.fittedvalues  
df
# plot = df[['Close', 'pronostico']].plot(figsize=(10, 8)) 
```
